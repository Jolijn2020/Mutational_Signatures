```python
hamming_dist = dict()
for m in mutations:
    hamming_dist[m] = dict()
    for j in mutations:
        hamming_dist[m][j] = distance.hamming(list(m), list(j)) * len(j)
            
            
```

```python
df = pd.DataFrame(hamming_dist).astype(int)
```

```python
df.to_csv('hamming_distances.tsv', sep='\t', index=True, header=True)
```

# Normalization steps:
- average within each cancer type
- then min-max normalize within each cancer type (all results are positive, and weighs against what is "much" within each type) 
- then average across all cancer types
- output per cancer type and total


```python
mutation_counts =  mutation_counts.T
mutation_counts = mutation_counts.assign(cancer_type=mutation_counts.index.to_series().str.split("::").str[0])
mutation_counts_means = mutation_counts.groupby('cancer_type').mean()
```

```python
mutation_counts_means
```

```python
means_normalized = (mutation_counts_means.T - mutation_counts_means.min(axis=1)) / (mutation_counts_means.max(axis=1) - mutation_counts_means.min(axis=1)).T
```

```python
means_overall = means_normalized.T.mean()
```

```python
diffs = dict()
for col_label, col in means_normalized.items():
    diffs[col_label] = dict()
    for index, value in col.items():
        diffs[col_label][index] = dict()
        for i2, v2 in col.items():
            diffs[col_label][index][i2] = abs(value-v2)
```

```python
AML = pd.DataFrame(diffs['AML']).sort_index().sort_index(axis=1)
```
